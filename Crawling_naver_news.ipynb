```python
def wordize(df):
  title = []
  contents = []
  tagger = Kkma()
  for x,y in tzip(df['뉴스타이틀'], df['내용']):
    tmp_x = tagger.nouns(x)
    tmp_y = tagger.nouns(y)
    title.append(tmp_x)
    contents.append(tmp_y)

  df_data['타이틀_word'] = title
  df_data['내용_word'] = contents
  return df_data
```

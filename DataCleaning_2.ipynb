```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("2021_기본조사_20221123_82663.csv",encoding='cp949')
data
```

```python
data.columns = ['장애인고용여부','장애인고용률','업종','본사소재지역','종사자규모','고용의무기업체여부','부담금납부대상여부','조직형태','전체근로자수(전체)','전체근로자수(남자)','전체근로자수(여자)','장애인근로자수(전체)','장애인근로자수(남자)','장애인근로자수(여자)','장애인(합계)','지체','뇌병변','시각','청각','언어','지적','정신','자폐성','신장','심장','호흡기','간','안면','장루요루','뇌전증','유공자']
data
```

```python
data['장애인고용여부'].replace(1,"장애인 고용기업체",inplace=True)
data['장애인고용여부'].replace(2,"장애인 미고용기업체",inplace=True)
```

```python
data['업종'].replace({1:"제조업",2:"서비스업1",3:"서비스업2",4:"기타산업"},inplace=True)
data['본사소재지역'].replace({1:"수도권",2:"광역시권",3:"기타지역"},inplace=True)
data['종사자규모'].replace({1:"1~4명",2:"5~49명",3:"50~299명",4:"300~999명",5:"1,000명 이상"},inplace=True)
data['고용의무기업체여부'].replace({1:"고용 비의무",2:"고용 의무"},inplace=True)
data['부담금납부대상여부'].replace({1:"부담금 납부 대상(100명 이상)",2:"부담금 납부 미대상(100명 미만)"},inplace=True)
data['조직형태'].replace({1:"개인기업체",2:"회사법인",3:"회사이외법인"},inplace=True)
```

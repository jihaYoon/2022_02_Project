```python
import pandas as pd
import numpy as np
```

```python
data2 = pd.read_csv("한국장애인고용공단_장애인 취업 정보(20210101_20211231).CSV",encoding='cp949')
data2
```

```python
data3 = pd.read_csv("한국장애인고용공단_장애인 취업 정보_20201231.csv",encoding='cp949')
data3
```

```python
data3 = data3.append(data2,sort=False, ignore_index=True)
data3
```

```python
df = data3['근무지역'].str.split(' ',expand=True)
df.columns=['근무지역(시도)','근무지역(시군구)','근무지역(구)','근무지역_null']
df = df.replace({'경북':'경상북도','대구':'대구광역시','강원':'강원도','경남':'경상남도','인천':'인천광역시',
                 '경기':'경기도','전남':'전라남도','충북':'충청북도','전북':'전라북도','충남':'충청남도',
                 '서울':'서울특별시','대전':'대전광역시','광주':'광주광역시','부산':'부산광역시','울산':'울산광역시','제주':'제주특별자치도'})

df = df.replace({'당진군':'당진시','마산시':'창원시','북제주군':'제주시','진해시':'창원시','청원군':'청주시','화성시(수원)':'화성시','화성시(화성)':'화성시'})
df['근무지역(시도)'].unique()
```

```python
data3 = data3.join(df)
data3
```

```python
data3 = data3.dropna(axis='index',how='any',subset=['근무지역(시도)'])
data3['근무지역(시도)'].unique()
data3
```

```python
a = data3.index[data3['근무지역(시도)'] == '세종특별자치시'].tolist()
for i,idx in enumerate(a):
  data3.loc[idx,'근무지역(시군구)'] = '세종특별자치시'

data3
```

```python
data3.to_csv("f$한국장애인고용공단_장애인 취업 정보(2020-2021).csv")
```
